```python
languages={
  'bangla':'0',
  'malayalam':'1',
  'odiya':'2',
  'telugu':'3'
}
observed_language=['bangla', 'malayalam', 'odiya', 'telugu']
```

```python
lst = []
import time
start_time = time.time()
c=0
path='accentdb_core/data/'
for subdir, dirs, files in os.walk(path):
    for file in files:
        #Load librosa array, obtain mfcss, store the file and the mcss information in a new array
        X, sample_rate = librosa.load(os.path.join(subdir,file), res_type='kaiser_fast')
        mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
        file_name=os.path.basename(file)   
        language=file_name.split("_")[0]
        #print(language)
        c+=1
        value_language=languages[language]
        print(value_language)
        
        arr = mfccs, value_language
        lst.append(arr)
        if ((c>=1000) and (c%1000==0)):
            print("this is to see how many files have been processed",c)
print("--- Data loaded. Loading time: %s seconds ---" % (time.time() - start_time))
```

```python
# Creating X and y: zip makes a list of all the first elements, and a list of all the second elements.
X, y = zip(*lst)
```

```python
X = np.asarray(X)
y = np.asarray(y)

X.shape, y.shape
```

```python
# Saving joblib files to not load them again with the loop above

X_name = 'X.joblib'
y_name = 'y.joblib'
save_dir = ''

savedX = joblib.dump(X, os.path.join(save_dir, X_name))
savedy = joblib.dump(y, os.path.join(save_dir, y_name))
```

```python
# Loading saved models

X = joblib.load('X.joblib')
y = joblib.load('y.joblib')
```

```python
#divide data test and train
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
x_traincnn = np.expand_dims(X_train, axis=2)
x_testcnn = np.expand_dims(X_test, axis=2)
```

```python
x_traincnn.shape, x_testcnn.shape
```

```python
#model creation in whic data will be fit
import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.utils import to_categorical
from keras.layers import Input, Flatten, Dropout, Activation
from keras.layers import Conv1D, MaxPooling1D
from keras.models import Model
from keras.callbacks import ModelCheckpoint

model = Sequential()

model.add(Conv1D(64, 5,padding='same',
                 input_shape=(40,1)))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(MaxPooling1D(pool_size=(4)))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(MaxPooling1D(pool_size=(4)))
model.add(Conv1D(256, 5,padding='same',))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(Flatten())
model.add(Dense(units=8,kernel_initializer='he_normal',Activation='relu')
model.add(Dense(4))
model.add(Activation('softmax'))
opt = keras.optimizers.RMSprop(lr=0.00005, rho=0.9, epsilon=1e-07, decay=0.0)
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
cnnhistory=model.fit(x_traincnn, y_train.astype(float), batch_size=16, epochs=200, validation_data=(x_testcnn, y_test.astype(float)))
```

```python
plt.plot(cnnhistory.history['loss'])
plt.plot(cnnhistory.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(cnnhistory.history['accuracy'])
plt.plot(cnnhistory.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
predictions = model.predict_classes(x_testcnn)
```
